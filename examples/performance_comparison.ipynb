```python
# Botorch and Ax imports.
from ax import json_load
from ax.modelbridge.registry import Models
from ax.models.torch.botorch_modular.surrogate import Surrogate
from ax.core.observation import ObservationFeatures
from botorch.models.gp_regression import SingleTaskGP
from botorch.acquisition.monte_carlo import qNoisyExpectedImprovement
from ax.service.utils.report_utils import exp_to_df

# Other library imports.
import matplotlib.pyplot as plt
import numpy as np
import scipy
import os
```

```python
# Path to BoTorch experiments.
exp_fps = [
    r"C:\Users\ay434\Box\Research\Flood_Sim_Materials\BayesOpt_Paper\Data\BT_exp_campagna_1veg_6sens.json",
    r"C:\Users\ay434\Box\Research\Flood_Sim_Materials\BayesOpt_Paper\Data\BT_exp_campagna_2veg_6sens.json"
]

# Legend labels for the models.
labels = [
    '1 Veg. LC',
    '2 Veg. LC'
]

# Where to save the figure.
fig_savedir = r"C:\Users\ay434\Box\Research\Flood_Sim_Materials\BayesOpt_Paper\Figures\Case_Study_1"
fig_savename = "iteration_compare.png"
fig_fp = os.path.join(fig_savedir, fig_savename)

# Load the experiments.
exps = [json_load.load_experiment(fp) for fp in exp_fps]

# Make list of dfs.

results_list = [exp_to_df(exp) for exp in exps]
```

```python
# Plot the performance over each iteration.
fig, ax = plt.subplots()
for c, df in enumerate(results_list):
    best_func = []
    for i in range(df.shape[0]):
        # Get the first i rows of the results.
        sample_df = df.iloc[:i+1,:]

        # Maximum function value of the sample rows.
        sort_sample_df = sample_df.sort_values(by='_run_with_loss', ascending=False)
        sort_sample_df.reset_index(inplace=True)
        best_func.append(sort_sample_df.loc[0, '_run_with_loss'])

    ax.plot(best_func, '-o', label=labels[c])
    ax.set_xlabel('Iteration Number', fontsize=18)
    ax.set_ylabel('$\overline{RMSE}$', fontsize=18)
    ax.tick_params(axis='x', labelsize=18)
    ax.tick_params(axis='y', labelsize=18)
    
ax.legend(fontsize=18)
plt.tight_layout()
fig.savefig(fig_fp, dpi=300)
```
